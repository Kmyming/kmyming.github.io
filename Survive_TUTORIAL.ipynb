```python
import pdb
from game.agent_ai import SimpleAI
from game.commands import Command, AttackCommand, TakeItemCommand, MoveCommand, IdleCommand, UseItemCommand

class PlayerOneAI(SimpleAI):
  @staticmethod
  def decide(state, prev_state=None):
    if INTERACTIVE_MODE:
      pdb.set_trace()
    items_that_i_can_move_to = state['indirect_takables']
    for row in items_that_i_can_move_to:
      item = row['item']
      shortest_path = row['shortest_path']
      if item['name']=='strawberry':
        return MoveCommand(shortest_path[0])
    items_on_my_cell = state['takables']
    for row in items_on_my_cell:
      item = row
      if item['name']=='strawberry':
        strawberry = item
        return TakeItemCommand(strawberry)
    items_on_my_inventory = state['self']['inventory']
    for row in items_on_my_inventory:
      item = row
      if item['name'] == 'strawberry':
        return UseItemCommand(item)
    items_that_i_can_move_to = state['indirect_takables']
    for row in items_that_i_can_move_to:
      item = row['item']
      shortest_path = row['shortest_path']
      if item['name']=='greatsword':
        return MoveCommand(shortest_path[0])
    items_on_my_cell = state['takables']
    for row in items_on_my_cell:
      item = row
      if item['name']=='greatsword':
        greatsword = item
        return TakeItemCommand(greatsword)
    monsters_or_players = state['indirect_attackables']
    for row in monsters_or_players:
      entity = row['agent']
      shortest_path = row['shortest_path']
      if entity['name']=='monster_M0':
        shortest_path_to_monster = shortest_path
        return MoveCommand(shortest_path_to_monster[0])
    nearby_monsters_or_players = state['direct_attackables']
    for row in nearby_monsters_or_players:
      entity = row
      if entity['name']=='monster_M0':
        return AttackCommand(entity)  
    return IdleCommand()
```
